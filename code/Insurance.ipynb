```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
# insurance
# candy
```

```python
df_insurance = pd.read_csv('insurance.csv')
```

```python
df_insurance.head()
```

```python
# scatter plot to get a relationship 

plt.figure(figsize=(6,4))
plt.title('Relationship between BMI and Charges')
sns.scatterplot(x= df_insurance['bmi'], y = df_insurance['charges'])
```

```python
sns.regplot(x = df_insurance['bmi'], y= df_insurance['charges'])
```

```python
sns.lmplot(x= 'bmi',y= 'charges', hue= 'smoker', data=df_insurance)
```

```python
sns.scatterplot(x= 'bmi',y= 'charges', hue= 'smoker', data=df_insurance)
```

```python
# categorical scatterplot 
sns.swarmplot(y = df_insurance['charges'],
             x = df_insurance['smoker'])
```
