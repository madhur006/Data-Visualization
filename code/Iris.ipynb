```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df_iris = pd.read_csv('iris.csv', index_col='Id')
```

```python
df_iris.head()
```

```python
# histogram numerical one column 
plt.figure(figsize=(4,4))
plt.title('petal lengths in flowers')
sns.histplot(data=df_iris['Petal Length (cm)'])
```

```python
sns.kdeplot(data=df_iris['Petal Length (cm)'], shade=True)
```

```python
sns.jointplot(x= df_iris['Petal Length (cm)'], y = df_iris['Sepal Width (cm)'], kind= 'kde')
```

```python
# color coded plots 

sns.histplot(x = df_iris['Petal Length (cm)'], hue= df_iris['Species'])
```

```python
sns.kdeplot(x = df_iris['Petal Length (cm)'], hue= df_iris['Species'], shade=True)
```
